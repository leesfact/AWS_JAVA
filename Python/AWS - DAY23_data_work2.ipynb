```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\HMFMPYUN.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
#print(kfont)
matplotlib.rcParams['font.family'] = [kfont]+current_font_list


# 데이터 불러오기
pd_data = pd.read_csv('데이터산업데이터직무인력현황.csv', encoding='cp949')
print(pd_data.head())
```

```python
fig,axes = plt.subplots(2,4, figsize = (6.4*4, 4.8*2), sharey = True, sharex=  True)
col_names = pd_data.columns[-2:]
for i, n in enumerate (pd_data['데이터직무별(1)'].unique()):
    filter1 = pd_data['데이터직무별(1)'] == n
    xs = pd_data.loc[filter1,'시점']
    ys = pd_data.loc[filter1,col_names]
    ax = axes[i//4,i%4]
    ax.set_title(n)
    ax.plot(xs.astype('str'),ys.iloc[:,0],'gv--')
    ax.plot(xs.astype('str'),ys.iloc[:,1],'ro--')
    #print(xs.dtype)
```

```python
filter1 = pd_data['데이터직무별(1)'] =='데이터 아키텍트'
col_names = pd_data.columns[-2:]
xs = pd_data.loc[filter1,'시점']
ys = pd_data.loc[filter1,col_names]
_,axe = plt.subplots()
axe.plot(xs,ys)
```
