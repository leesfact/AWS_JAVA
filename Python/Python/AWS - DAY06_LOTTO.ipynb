```python
import math
def mean(num_list):
    return (sum(num_list)/len(num_list))  

def dev(num_list): #편차
    m = mean(num_list)
    return [x-m for x in num_list]
    
def var(num_list): # 분산
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d]) / (n-1)

def stdev(num_list): #표준편차
    return math.sqrt(var(num_list))   
```

```python
csv_data = []
with open('lott.csv',encoding='cp949') as f:
    for line in f:
        #print(line)
        csv_data.append(line[:-1].split(','))
        #csv_data.append(line.replace('\n','').split(','))
#print(csv_data[:10])
print(csv_data)
#int(csv_data[0][-1])
```

```python
#1 . 전체 회차를 통틀어 빈도수 (mode, 최빈값): 숫자별 카운트 1-45
for game in csv_data: # 헤더 없음
    for i,num in enumerate(game) :
        game[i] = int(num)
print(csv_data[:3]) #3번째 까지 확인용
```

```python
for i in range(len(csv_data)):
    for j in range(len(csv_data[i])):
        csv_data[i][j] = int(csv_data[i][j])
print(csv_data[:3])
```

```python
#flat
lott_num_flat = []
for game in csv_data:
    lott_num_flat = lott_num_flat + game
print(lott_num_flat[:20])
```

```python
#fla2
lott_num_flat2 = []
for i in range(len(csv_data)):
    for j in range(len(csv_data[i])):
        lott_num_flat2.append(csv_data[i][j])
print(lott_num_flat2[:20])
```

```python
lott_num_domain = list(set(lott_num_flat))
lott_num_freq = []
for num in lott_num_domain:
    x = lott_num_flat.count(num)
    lott_num_freq.append(x)
#print(lott_num_domain[:45])
#print(lott_num_freq[:45])
dict_freq = dict(zip(lott_num_domain,lott_num_freq))
dict_freq = dict(sorted(dict_freq.items(),key = lambda x: x[1], reverse=True))
#print(dict_freq)

x = list(dict_freq.keys())
print(x[:6])

# mode = 43
```

```python
#2. 회차별 - 평균값, 표준편차

for i,game in enumerate(csv_data): #game = list
    print('회차: {}, \t평균: {} \t표준편차:{}'.format(i+1, mean(game),stdev(game)))
```
