```python
import math
def mean(num_list):
    return (sum(num_list)/len(num_list))  

def dev(num_list): #편차
    m = mean(num_list)
    return [x-m for x in num_list]
    
def var(num_list): # 분산
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d]) / (n-1)

def stdev(num_list): #표준편차
    return math.sqrt(var(num_list))   

# 공분산
def covar(list_1, list_2): #공분산
    n = len(list_1)
    list_1_dev = dev(list_1)
    list_2_dev = dev(list_2)
    
    return sum(x * y for x, y in zip(list_1_dev,list_2_dev))/(n-1)

# 상관도
def corel(list_1, list_2): # 상관도 (-1 ~ 1)
    return covar(list_1,list_2) / (stdev(list_1)*stdev(list_2))
    
```

```python
import requests
link = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data')

print(link.text)
```

```python
pass1 = link.text.split('\n') # by - line
#print(len(pass1))  #데이트시트 차종의 갯수
#print(pass1[:1])

pass2 = [] # 각 샘플(줄) 별 숫자부분, 차명 부분 분리, split('\t')

for line in pass1:
    pass2.append(line.split('\t'))
pass3 = []  # 숫자 부분 분리. split()
for e in pass2:
    try:
        y = e[0].split()
        y.append(e[1])
        pass3.append(y)
    except:
        pass
print(pass3)

```

```python
   # float형으로 형변환, 차명 "제거

for e in pass3 :
    for i, n in enumerate(e[:-1]):
        try:
            e[i] = float(n)
        except:
            print(n)  # 결측지 확인 ## print(1) 결측지 처리
            e[i] = 0.0
    e[-1] = e[-1].strip('"')


```

```python
pass4 = []
#print(pass3)

mpg_list = []
disp_list = []

for e in pass3 : 
    mpg_list.append(e[0])   # 연비
    disp_list.append(e[2])  # 배기량

print(mpg_list)
print(disp_list)
            
```

```python
# 연비 : 평균, 표준편차
print('평균: ',mean(mpg_list))
print('표준편차: ',stdev(mpg_list))


# 배기량 : 평균, 표준편차
print('평균: ', mean(disp_list)) 
print('표준편차: ', stdev(disp_list))


```

```python
print('상관도: ',corel(mpg_list, disp_list))
```
