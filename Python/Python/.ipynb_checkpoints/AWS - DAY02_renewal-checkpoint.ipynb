```python
#f = open('전국건강증진센터표준데이터.csv')
#print(f)
data_list = []
with open('전국건강증진센터표준데이터.csv',encoding ='cp949')as f:
    for line in f :
        #print(line,type(line))
        #print(line.split(','))
        data_list.append(line.split(','))
#data_list = data_list[1:]
print(data_list[:3])
```

```python
city_list = []
for i in data_list[1:]:
    if len(i[2])>0:
        city_list.append(i[2].split()[0])
        #print(i[2].split()[0])
       # print(i[2].split())
print(city_list)
    
  
```

```python
for n in city_list:
    if n[0] =='"':
        print(n[1:])
    else:
        print(n)
```

```python
for i,n in enumerate(city_list):
    city_list[i] = n.lstrip('"')
    print(city_list[:10])
```

```python
for n in set(city_list):
    print(n, city_list.count(n))
```

```python
#빈도수 dictionary 반환
#1. raw_list를 set으로 변환
#2. set을 순환하면서, 각각 몇개씩 raw_list에 있는지 카운트
#3. 1,2에서 얻은 key, value 값으로 dictionary 만들어 return

# lamdba x: x[1]
def sort_key(x):
    return x[1]




def unique_count(raw_list, sort=False):
    raw_set = set(raw_list)
    k_list = []
    v_list = []
    for i in raw_set:
        #print(i,raw_list.count(i))
        k_list.append(i)
        v_list.append(raw_list.count(i))
   # print(d_raw_list)
   # print(d_raw_count)
    ret_dict = dict(zip(k_list,v_list)) 
    if sort == True:
        
        ret_dict = dict(sorted(ret_dict.items(),key=lambda x: x[1]))
    return ret_dict
    
unique_count(city_list, sort=True)
```
