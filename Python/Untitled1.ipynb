```python
data_list=[]

with open('전국.csv',encoding = 'EUC-KR') as f: #module as 별칭
    for line in f:
        #print(line.split(','))
        data_list.append(line.split(','))
        print(data_list[:3])
```

```python
city_list = []

for i in data_list[1:]: # 헤더 짜름
    #print(i[2]) # i[2]는 건강센터 주소를 의미
    if len(i[2]) > 0: # 빈칸이 없으면
        city_list.append(i[2].split()[0])
        #print(i[2].split()[0]) #[0] : 도,시 [1] : 구 [2] : 길
print(city_list) # 도, 시 단위로 다 짜름
        
    
```

```python
for n in city_list:
 if n[0] =='"':
        print(n[1:])
else:
        print(n)
```

```python
for i ,n in enumerate(city_list) :
    city_list[i] = n.lstrip('"')
    print(city_list[:10])
```

```python
for n in set(city_list):
    print(n,city_list.count(n))  #n = 도, 시 이름
```

```python


def sort_key(x):
    return x[1]
def unique_count(raw_list,sort = False):
    raw_set=set(raw_list)
    key_list = []
    value_list = []
    for i in raw_set:
        #print(i, raw_list.count(i))
        key_list.append(i)
        value_list.append(raw_list.count(i))
   # print(key_list)
   # print(value_list)
    ret_dict = dict(zip(key_list,value_list))
    
    if sort == True:
        ret_dict = dict(ret_dict.items(),key=lambda x:x[1])
    return ret_dict


unique_count(city_list,sort = True)
```
