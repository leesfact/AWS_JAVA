```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
xs = np.linspace(0, 50, 100)
ys = np.sin(xs)

#print(xs)
#print(ys)

figure, axe = plt.subplots()
figure.set_facecolor('yellow') # 배경에 색 입히기 
figure.suptitle('Figure1')
figure.supxlabel('Figure1_x')
figure.supylabel('Figure1_y')
figure.text(0.5,0.7,'mytext')
figure.set_size_inches(15,4.8)
axe.plot(xs,ys)
axe.set_title('my axe title')
axe.set_xlabel('axe_x')
axe.set_ylabel('axe_y')
axe.set_ylim(0,1)
axe.set_xlim(0,30)
axe.set_xticks(np.linspace(0,30,5))
axe.set_xticklabels('abcde')
print(type(figure))
print(type(axe))
```

```python
fig2 = plt.figure(figsize=(15,7))
a = fig2.add_subplot(2,1,1)
b = fig2.add_subplot(2,2,3)
c = fig2.add_subplot(2,2,4)
```

```python
#이 코드는 figsize가 (15,7)이고 subplot이 2x2인 Figure 객체를 생성합니다.

#첫 번째 subplot은 2x1 행렬의 첫 번째 칸(즉, 1행)에 위치하며, 두 번째 subplot은 2x2 행렬의 세 번째 칸(즉, 2행 1열)에 위치하며, 세 번째 subplot은 2x2 행렬의 네 번째 칸(즉, 2행 2열)에 위치합니다.

#따라서 첫 번째 subplot은 a 변수에 할당되며, 두 번째 subplot은 b 변수에 할당되고, 세 번째 subplot은 c 변수에 할당됩니다.
fig3 = plt.figure(figsize=(15,7))
a = fig3.add_subplot(3,2,1)
b = fig3.add_subplot(3,2,2)
c = fig3.add_subplot(3,1,2)
d = fig3.add_subplot(3,2,5)
e = fig3.add_subplot(3,2,6)

fig3.savefig('fig3.png')
```
