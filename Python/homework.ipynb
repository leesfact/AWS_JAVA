```python
import numpy as np
import math

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

def split_len(data_list):
    len_list=[]
    for e in data_list:
        len_list.append(len(e))
    print(set(len_list))
    if len(set(len_list))>1:
        for i in set(len_list):
            print(i, len_list.count(i))
```

```python
csv_data = []
with open('전국로컬푸드인증정보표준데이터.csv',encoding='EUC-KR') as f:
    for line in f:
         #print(line)
        csv_data.append(my_split(line))
        
print(csv_data[:3])

```

```python
p1 = []
csv_data = csv_data[1:]
for i in csv_data:
    p1.append(i)

print(p1[:3])
```

```python


p1 = [[e for e in sublist if e != ''] for sublist in p1]

print(p1[:5])


```

```python
p2 = []
for e in p1:
    y = e[3].split()[:3]
    y.append(e[5])
    p2.append(y)
print(p2[:50])


```

```python

p3 = []
for s in p2:
    for i, e in enumerate(s):
        s[i] = e.replace("+", ",")
    p3.append(s)
print(p3[:20])
```

```python
print(p3[:3])

```

```python
counts = {}

for row in p3:
    if len(row) >= 4:
        value = row[3]
        if value in counts:
            counts[value] += 1
        else:
            counts[value] = 1

print(counts)
```
