```python
list_a = [1,2,3]     # key_list
list_b = list('abc') # value_list
d1 = {'a':1,'b':2,'c':3}
print(d1)

d2 = dict(zip(list_b,list_a))  #zipper
print(d2)
print(d2.items())

d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d3)

print(d3['c'])
print(d3.get('c'))
print(d3.get('x'))
d3['y'] = 15
print(d3)
```

```python
for i in d3:
    print(i,d3[i])
print(d3.keys())
print(d3.values())

for i in d3.values():
        print(i)
        
print(d3)
d3['y'] = 20
print(d3)
```
