# 데이터산업의 기술 등급별 데이터 직무 인력 현황

- 데이터 직무별 axe 생성
    - x축 : 시점
    - y축 : 업종(2개)
- 하나의 figure에, 총 8개 axe가 바둑판 형식으로 배치되는 형태 (4 col X 2 rows) ,figure 사이즈는 각자 알아서 
- title 표시

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python

# 데이터 불러오기
pd_data = pd.read_csv('데이터산업데이터직무인력현황.csv', encoding='cp949')

current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\HMFMPYUN.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
#print(kfont)
matplotlib.rcParams['font.family'] = [kfont]+current_font_list
```

```python

pd_data = pd_data[['데이터직무별(1)', '시점', '데이터 처리 및 관리 솔루션 개발·공급업', '데이터 구축 및 컨설팅 서비스업']]
pd_data = pd_data.pivot_table(index='시점', 
                              columns=['데이터 처리 및 관리 솔루션 개발·공급업', '데이터 구축 및 컨설팅 서비스업'], values='데이터직무별(1)')

# figure와 axe 생성
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.ravel()

# 데이터 직무별 시각화
for i, col in enumerate(pd_data.columns):
    axe = axes[i]
    axe.plot(pd_data.index, pd_data[col])
    axe.set_title(col)
    axe.set_xlabel('시점')
    axe.set_ylabel('데이터 직무별 인력')
    ymin, ymax = pd_data[col].min(), pd_data[col].max()
    margin = (ymax - ymin) * 0.1  # 상하 여백 10% 확보
    axe.set_ylim([ymin - margin, ymax + margin])
    
# 전체 그래프의 타이틀
plt.suptitle('데이터산업의 기술 등급별 데이터 직무 인력 현황', fontsize=20)

# 그래프 간 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
