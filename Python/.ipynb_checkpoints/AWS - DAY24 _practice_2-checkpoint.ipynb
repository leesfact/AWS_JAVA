```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\HMFMPYUN.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
#print(kfont)
matplotlib.rcParams['font.family'] = [kfont]+current_font_list

pd_data = pd.read_csv('데이터산업시장규모.csv', encoding='cp949')
print(pd_data.head())
```

```python
fig = plt.figure(figsize=(16, 8), tight_layout=True)

line_axe = fig.add_subplot(2,1,1)
col_names = pd_data.columns
line_params = ['ro-','g^--','bv:']
line_axe.set(title = '2015~2021 산업별 규모',ylabel = '억원')

for i in range(1,4):
    line_axe.plot('시점',col_names[i], line_params[i-1],data = pd_data, label=col_names[i])

line_axe.legend()

pie2015_axe = fig.add_subplot(2,2,3)
pie2015_axe.set_title('2015년 구성비', y = -0.3)
pie2015_axe.pie(pd_data.iloc[0, 1:],labels = pd_data.columns[1:],autopct='%1.1f%%')
pie2021_axe = fig.add_subplot(2,2,4)
pie2021_axe.set_title('2021년 구성비', y = -0.3)
pie2021_axe.pie(pd_data.iloc[-1, 1:],labels = pd_data.columns[1:],autopct='%1.1f%%')

plt.suptitle('2015년과 2021년 데이터 산업별 구성비 및 시장규모 성장 추이', fontsize=14, fontweight='bold')
plt.show()
```
