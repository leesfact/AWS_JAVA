```python
import numpy as np
import math

def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list

def split_len(data_list):
    len_list=[]
    for e in data_list:
        len_list.append(len(e))
    print(set(len_list))
    if len(set(len_list))>1:
        for i in set(len_list):
            print(i, len_list.count(i))
```

```python
csv_data = []
with open('전국로컬푸드인증정보표준데이터.csv',encoding='EUC-KR') as f:
    for line in f:
         #print(line)
        csv_data.append(my_split(line))
        
print(csv_data[:3])
```

```python
p1 = []
csv_data = csv_data[1:]
for i in csv_data:
    p1.append(i)

print(p1[:3])
```

```python
p2 = []
for e in p1:
    y = e[15].split()
    y.append(e[8])
    p2.append(y)
print(p2[:20])
```

```python
p3 = []
for s in p2:
    for i, e in enumerate(s):
        s[i] = e.replace("+", ",")
    p3.append(s)
print(p3[:20])
```

```python
p4 = []
for i in p3:
    if len(i) >= 3:
        x = i[0] + ' ' + i[1]
        y = i[2]
        p4.append([x, y])
#print(p4)
```

```python
v = np.array(p4)
print(v[:10])
```

```python
v2 = []
for i in v:
    if ',' in i[1]:
        for p in i[1].split(','):
            v2.append([i[0], p])
    else:
        v2.append(i)

print(v2[:10])
```

```python
v3 = v2

v3 = [x.tolist() if isinstance(x, np.ndarray) else x for x in v2]

# 리스트인 data의 각 요소에 대해서 isinstance 함수를 사용하여 
# 해당 요소가 np.ndarray 자료형인지 검사, np.ndarray 자료형이라면 tolist 메소드를 사용하여 
# 리스트로 변환

print(v3[:10])
```

```python
for row in v3:
    print(row)
```

```python
v3 = [[s.replace('"', '') for s in e] for e in v3]
for row in v3:
    print(row)
```

```python

result = {}
for item in v3:
    key = item[0]
    value = item[1]
    if key in result:
        result[key].append(value)
    else:
        result[key] = [value]
print(result)
```
