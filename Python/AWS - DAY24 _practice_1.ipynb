```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python

current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\HMFMPYUN.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
#print(kfont)
matplotlib.rcParams['font.family'] = [kfont]+current_font_list

pd_data = pd.read_csv('데이터산업시장규모.csv', encoding='cp949')
print(pd_data.head())
```

```python
# plot 그래프
plt.plot(pd_data['시점'], pd_data['데이터 처리 및 관리 솔루션 개발·공급업'], label='데이터 처리 및 관리 솔루션 개발·공급업')
plt.plot(pd_data['시점'], pd_data['데이터 구축 및 컨설팅 서비스업'], label='데이터 구축 및 컨설팅 서비스업')
plt.plot(pd_data['시점'], pd_data['데이터 판매 및 제공 서비스업'], label='데이터 판매 및 제공 서비스업')

# x, y 축 레이블 및 제목 추가
plt.xlabel('시점')
plt.ylabel('시장규모')
plt.title('시점별 데이터 직무 시장규모 성장 추이')

# 범례 추가
plt.legend()

# 그래프 출력
#plt.show()
```

```python
# pratice
fig = plt.figure() # figsize, layout
a = fig.add_subplot(2,1,1)
b = fig.add_subplot(2,2,3)
c = fig.add_subplot(2,2,4)
```

```python
fig, axes = plt.subplots(1, 2)

# 첫번째 subplot에 2015년 데이터를 그립니다.
axes[0].pie(pd_data.iloc[0, 1:], labels=pd_data.columns[1:], autopct='%1.1f%%')
axes[0].set_title('2015년')

# 두번째 subplot에 2021년 데이터를 그립니다.
axes[1].pie(pd_data.iloc[1, 1:], labels=pd_data.columns[1:], autopct='%1.1f%%')
axes[1].set_title('2021년')

# subplot 간 간격 조정
plt.subplots_adjust(wspace=0.5)

# 전체 그래프 제목
plt.suptitle('2015년과 2021년 데이터 산업별 구성비')

# 그래프 출력
plt.show()

```

```python
# plot 그래프
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

ax1.plot(pd_data['시점'], pd_data['데이터 처리 및 관리 솔루션 개발·공급업'], label='데이터 처리 및 관리 솔루션 개발·공급업')
ax1.plot(pd_data['시점'], pd_data['데이터 구축 및 컨설팅 서비스업'], label='데이터 구축 및 컨설팅 서비스업')
ax1.plot(pd_data['시점'], pd_data['데이터 판매 및 제공 서비스업'], label='데이터 판매 및 제공 서비스업')

# x, y 축 레이블 및 제목 추가
ax1.set_xlabel('시점')
ax1.set_ylabel('시장규모')
ax1.set_title('시점별 데이터 직무 시장규모 성장 추이')

# 범례 추가
ax1.legend()

# pie 그래프
ax2[0].pie(pd_data.iloc[0, 1:], labels=pd_data.columns[1:], autopct='%1.1f%%')
ax2[0].set_title('2015년')

ax2[1].pie(pd_data.iloc[1, 1:], labels=pd_data.columns[1:], autopct='%1.1f%%')
ax2[1].set_title('2021년')

# 전체 그래프 제목
plt.suptitle('2015년과 2021년 데이터 산업별 구성비 및 시장규모 성장 추이', fontsize=14, fontweight='bold')

# 그래프 출력
plt.show()
```
