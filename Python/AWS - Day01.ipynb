```python
# 100 아래 (미만) 숫자 중 3 또는 5로 나누어지는 숫자들의 합

list1 = [x for x in range(0,100,3)]
print(list1)

list2 = [x for x in range(0,100,5)]
print(list2)

print(list1 + list2)

s = set(list1+list2) 
print(s)
```

```python
list_3 = [i for i in range(0,100,3)]
list_5 = [i for i in range(0,100,5)]
print(list_3)
print(list_5)
print(list_3 + list_5)

list_all = []

for i in list_3:
    list_all.append(i)
for i in list_5:
    list_all.append(i)
print(list_all)

set1 = set(list_all)
print(set1)

s = 0

for i in set1:
    s = s+i
    print(s)
print(sum(set1))
```
